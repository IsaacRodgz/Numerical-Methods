## Newton interpolation

Importa numpy para manejo de arreglos, matplotlib para la graficación y funciones de interpolación escritas en C.<br><br>
Biblioteca interp1 contiene funciones:
* newton_interp: interpolación de Newton
* lagrange_interp: interpolación de Lagrange

```python
import matplotlib.pyplot as plt
import numpy as np
import interp1
```

Se definen los puntos $x_0, x_1, ..., x_n$ y los valores de $f(x_0), f(x_1), ..., f(x_n)$ para $f(x) = sin(x)$

```python
x = np.linspace(-7,7,10)
f = np.sin(x)
```

Se definen los valores de x en donde se quiere realizar la interpolación

```python
 p = np.array([-3.1415, -1, 1.57075, 6.5])
```

Se calculan los valores $y = f(x)$ para los puntos en p. También se calcula el timepo de ejecución.

```python
%time y = interp1.newton_interp(x,f,p,len(p))
```

Se compara la función $f(x)$ con los valores (p, y) interpolados mediante Newton

```python
plt.plot(p, y, 'ro')

xt = np.linspace(-7,7,100)
ft = np.sin(xt)

plt.plot(xt, ft)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

## Lagrange interpolation

Se calculan los valores $y = f(x)$ para los puntos en p. También se calcula el timepo de ejecución.

```python
%time y = interp1.lagrange_interp(x,f,p,len(p))
```

```python
plt.plot(p, y, 'ro')

xt = np.linspace(-7,7,100)
ft = np.sin(xt)

plt.plot(xt, ft)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

## Hermite interpolation

```python
x = np.linspace(-3.14,3.14,3)
f = np.sin(x)
fp = np.cos(x)
```

Se definen los valores de x en donde se quiere realizar la interpolación

```python
p = np.linspace(-3.14,3.14,100)
```

Se calculan los valores $y = f(x)$ para los puntos en p. También se calcula el timepo de ejecución.

```python
%time y = interp1.hermite_interp(x,f,fp,p,len(p))
```

```python
plt.plot(p, y, 'r')

xt = np.linspace(-3.14,3.14,100)
ft = np.sin(xt)

plt.plot(xt, ft)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```
